#### 3.3 Information about our dataset:<a class="anchor" id="information_about_our_dataset_3_3"></a>  

The dataset was created using survey responses collected via Amazon Mechanical Turk between December 3rd and December 5th, 2016. Thirty eligible Fitbit users voluntarily shared their personal tracker data, which includes minute-by-minute records of physical activity, heart rate, and sleep patterns. Variations in the data stem from differences in Fitbit device models and individual tracking habits.

#### 3.4 Verification and Organization of Data: <a class="anchor" id="verification_and_organization_of_data_3_4"></a>

The dataset consists of 18 CSV files, each capturing different quantitative metrics tracked by Fitbit. The data is in a "long" format, where each row represents a single time point for a specific user, with multiple entries per user as data is recorded daily and hourly. Each user is assigned a unique ID for identification.

Given the small sample size, I utilized pivot tables in Google Docs to sort and filter the data for a more structured analysis. This method allowed me to examine attributes and observations across tables, identify relationships between them, and confirm that the analysis spans a 31-day period.

### 5.1 Type of users per activity level <a class="anchor" id="type_of_users_per_activity_level"></a>


Since our dataset lacks demographic variables, we aim to classify users based on their activity levels using the available data. We will achieve this by categorizing users according to their daily step count, following these classifications:

* Sedentary – Fewer than 5,000 steps per day.
* Lightly active – Between 5,000 and 7,499 steps per day.
* Fairly active – Between 7,500 and 9,999 steps per day.
* Very active – More than 10,000 steps per day.
  
This classification is based on the guidelines from the article "Counting Steps: A Guide to Increasing Physical Activity Through Pedometer Use" (available at 10000steps.org.au). By adopting this approach, we can better understand user activity levels, uncover patterns and trends, and use these insights to refine our marketing strategy.

### 5.4 Correlations <a class="anchor" id="correlations"></a>

Certainly, we can examine whether there is a correlation between different variables. 

* Daily steps and daily sleep

* Daily steps and calories

Based on the analysis, there appears to be no significant correlation between daily activity levels, as measured by steps, and the amount of sleep users get each day. This suggests that the amount of physical activity does not influence sleep duration, indicating that these two variables are independent of each other.

In contrast, we observed a positive correlation between daily steps and calories burned. This implies that as the number of steps increases, the calories burned also tends to rise. This relationship is likely due to the fact that increased physical activity leads to greater energy expenditure, which consequently results in more calories burned. However, it's important to emphasize that while there is a correlation, it does not necessarily mean causation, and additional analysis would be needed to explore whether there is a direct cause-and-effect link between these two variables.

Here are some valuable recommendations based on our analysis:

**Daily Step Notifications and App Posts**: This recommendation aims to encourage users to meet the daily step goal of 8,000 steps as recommended by the CDC. We can send reminders to users who haven't reached this goal, and create posts within the app explaining the benefits of achieving this target. Additionally, we can highlight the positive correlation between step count and calorie burn. This approach will motivate users to be more active while educating them about the benefits of a healthy lifestyle.

**Reward System**: We propose developing an in-app game that motivates users to maintain consistent activity levels over time. The game would feature various levels based on the number of steps taken each day, with users earning stars for completing each level. These stars could then be redeemed for merchandise or discounts on other Bellabeat products. This strategy would not only encourage users to stay active but also reward their efforts in a fun and engaging way.

**Sleep Improvement Notifications and Techniques**: This recommendation involves offering users resources to improve their sleep quality. Users could set a desired bedtime, and receive notifications shortly before to help them prepare for sleep. Additionally, we could offer breathing exercises, relaxing podcasts, and sleep techniques to support better sleep habits. This initiative will enhance users' overall health and well-being.
